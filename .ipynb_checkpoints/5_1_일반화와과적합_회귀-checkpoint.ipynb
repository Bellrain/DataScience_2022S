## 데이터 생성

```python
import numpy as np
```

```python
real_X=np.random.uniform(size=100)
```

```python
real_y=4*real_X**2-5.5*real_X+4+0.2*np.random.normal(0.04, size=100)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(real_X,real_y)
```

```python
import pandas as pd
df=pd.DataFrame(np.column_stack((real_X,real_y)))
df.columns=['X','y']
```

```python
sample=df.sample(10)
plt.scatter(real_X,real_y)
plt.scatter(sample['X'], sample['y'])
```

```python
sample=sample.sort_values(by=['X'])
X=sample['X']
y=sample['y']
```

```python
X
```

## 1차회귀

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
X=np.array(X).reshape(-1,1)
reg.fit(X,y)
```

```python
y_pred=reg.predict(X)
```

```python
plt.scatter(real_X,real_y)
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y,y_pred)
```

## 2차항이 포함된 식으로 변경하기

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2,include_bias=False)
X_2=poly.fit_transform(X)
X_2
```

```python
reg.fit(X_2,y)
```

```python
y_pred=reg.predict(X_2)
```

```python
plt.scatter(real_X,real_y)
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
mean_squared_error(y,y_pred)
```

## 3차식

```python
poly = PolynomialFeatures(degree=3,include_bias=False)
X_temp=poly.fit_transform(X)
reg.fit(X_temp,y)
y_pred=reg.predict(X_temp)
mean_squared_error(y,y_pred)
```

```python
plt.scatter(real_X,real_y)
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
poly = PolynomialFeatures(degree=8,include_bias=False)
X_temp=poly.fit_transform(X)
reg.fit(X_temp,y)
y_pred=reg.predict(X_temp)
mean_squared_error(y,y_pred)
```
